```python
import mpy
import numpy as np
from mpy.specMagneticField import SPECField
from mpy.specMagneticField import traceLine, plotPoincare
```

```python
bField = SPECField(mpy.SPECOut("test.sp.h5"), lvol=0, sResolution=32, thetaResolution=32, zetaResolution=32)

nlines = 21
s0 = np.linspace(-0.99, 0.4, nlines)
theta0 = np.random.rand(nlines)
zeta0 = np.zeros(nlines)

lines = traceLine(
    bField=bField, s0 = s0, theta0 = theta0, zeta0 =zeta0,
    niter = 256, nstep = 32,
    bMethod = "calculate",
    jacobianData = "jacobianData_128.h5"
)

for i, line in enumerate(lines):
    line.writeH5("line"+str(i)+".h5")
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plotPoincare(lines, ax=ax)
plt.axis("equal")
```
